```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, 
                            n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100) #choosing all rows of column 1 nad 2
```

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1) #insert(array, column no. , value to be added, axis=column)
    weights = np.ones(X.shape[1]) #initiate weights as 1
    lr = 0.1 #learning rate
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights)) #dot product and then convert the final value to 0 or 1
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:] #returns bias, weights
```

```python
def step(z): #activation function
    return 1 if z>0 else 00
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1]) #Converts perceptron equation into line form:
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100) #creates 100 values between -3 and 3
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
