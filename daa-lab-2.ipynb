# **Bubble Sort**

```python

lis=[]

while True:
    ele = input("Enter an integer: ")

    if ele != "exit":
        lis.append(int(ele))
    else:
        break

print(f"Array before sort: {lis}")
#bubble sort
n = len(lis)
for i in range (n-1):
    for j in range (0,n-i-1):
        if lis[j]>lis[j+1]:
            lis[j], lis[j+1] = lis[j+1], lis[j]
    print(f"Array after iteration {i+1} : {lis}")

print(f"Array after sorting: {lis}")
```

# **Insertion Sort**

```python
lis=[]

while True:
    ele = input("Enter an integer: ")

    if ele != "exit":
        lis.append(int(ele))
    else:
        break

print(f"Array before sort: {lis}")

#insertion sort
n = len(lis)

for i in range(1,n):

    temp = lis[i]
    j=i-1

    while j>=0 and lis[j]>temp:
        lis[j+1] = lis[j]
        j -= 1

    lis[j+1] = temp
    print(f"Array after iteration {j+1}: {lis}")

print(f"Array after sorting: {lis}")
```

```python
lis = []

while True:
    ele = input("Enter an integer: ")

    if ele.lower() == "exit":
        break

    try:
        lis.append(int(ele))
    except ValueError:
        print("Please enter a valid integer")

print(f"Array before sort: {lis}")

# Insertion Sort
n = len(lis)

for i in range(1, n):
    key = lis[i]
    j = i - 1

    while j >= 0 and lis[j] > key:
        lis[j + 1] = lis[j]
        j -= 1

    lis[j + 1] = key
    

print(f"Array after sorting: {lis}")
```
